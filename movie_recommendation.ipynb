```python
# genres should be in the format of ['Action','Adventure','Fantasy','Science Fiction']
# to conver string of list to list we will use ast
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(1)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(obj):
    try:
        crew_list = ast.literal_eval(obj)
        directors = [member['name'] for member in crew_list if member.get('job') == 'Director']
        return directors
    except (SyntaxError, ValueError):
        # Handle the case where obj is not a well-formatted list
        return []
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer
```

```python
def stem(text):
    y = []
    for i in text.split():
        try:
            y.append(ps.stem(i))
        except TypeError:
            y.append(i)  # If there's a TypeError, simply append the original word
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)

```
